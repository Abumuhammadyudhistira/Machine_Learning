# K Means Clustering

Unsupervised Learning

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
# Create data sample, feature, center, cluster std
data = make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.8,random_state=101)
```

```python
data # numpy array [0] point, [1] cluster
```

```python
data[0].shape # row x column
```

```python
# Visualize Data
sns.set()
plt.scatter(data[0][:,0],data[0][:,1])
# all the row in the 0 column and 1 column
```

```python
data[1]
```

```python
sns.set()
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

```python
# Creating cluster
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4) # create cluster
```

```python
kmeans.fit(data[0]) # fit only to data[0]
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))

ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')

ax2.set_title("Original")
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

```python
# we try for 2 cluster
kmeans = KMeans(n_clusters=2)
kmeans.fit(data[0])

fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))

ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')

ax2.set_title("Original")
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```
